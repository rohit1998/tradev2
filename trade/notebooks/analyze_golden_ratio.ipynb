```python
import pandas as pd
import os
this_dir = os.path.abspath('')
this_dir
```

```python
df = pd.read_csv(os.path.join(this_dir, '../../output/golden_strat_test_full/0_10.csv'))
df.head()
```

```python
df.sort_values(by='p/l', ascending=False)
```

```python
newf = df[['sma','fma','p/l']]
newf = newf.pivot(index='sma', columns='fma', values='p/l')
newf
```

```python
import numpy as np 
import seaborn as sns
%matplotlib inline
sns.set(rc={'axes.facecolor':'pink', 'figure.facecolor':'pink'})
cmap = sns.diverging_palette(65, 180, s=100, l=80,n=100, center='light',as_cmap=False)
sns.heatmap(newf, annot=False,cmap=cmap)
```

```python
sns.heatmap(newf, annot=False)
```
